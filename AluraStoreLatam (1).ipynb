### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Ingreso total de los precios de cada tienda
tienda_total1 = tienda["Precio"].sum()
print(f'El total de los precios en la primer tienda es {tienda_total1}')

tienda_total2 = tienda2["Precio"].sum()
print(f'El total de los precios en la primer tienda es {tienda_total2}')

tienda_total3 = tienda3["Precio"].sum()
print(f'El total de los precios en la primer tienda es {tienda_total3}')

tienda_total4 = tienda4["Precio"].sum()
print(f'El total de los precios en la primer tienda es {tienda_total4}')
```

# 2. Ventas por categoría

###calcular la cantidad de productos vendidos por categoría en cada tienda.

```python
'''
Para calcular la cantidad de ventas por cada categoría en cada tienda es necesario
Agrupar agrupar las categorías por tienda y hacer la suma de la cantidad de coutas
que vendio cada producto en la tienda
'''
#Tienda 1
Ventas_Categoria = tienda.groupby('Categoría del Producto')['Cantidad de cuotas'].sum()
Ventas_Categoria.head(10)
```

```python
#Tienda 2
Ventas_Categoria2 = tienda2.groupby('Categoría del Producto')['Cantidad de cuotas'].sum()
Ventas_Categoria2.head(10)
```

```python
#Tienda 3
Ventas_Categoria3 = tienda3.groupby('Categoría del Producto')['Cantidad de cuotas'].sum()
Ventas_Categoria3.head(10)
```

```python
#Tienda 4
Ventas_Categoria4 = tienda4.groupby('Categoría del Producto')['Cantidad de cuotas'].sum()
Ventas_Categoria4.head(10)
```

# 3. Calificación promedio de la tienda


```python
calificacion_Promedio1 = tienda['Calificación'].mean()
print(f'Calificación promedio de la primer tienda {calificacion_Promedio1}')

calificacion_Promedio2 = tienda2['Calificación'].mean()
print(f'Calificación promedio de la segunda tienda {calificacion_Promedio2}')

calificacion_Promedio3 = tienda3['Calificación'].mean()
print(f'Calificación promedio de la tercera tienda {calificacion_Promedio3}')

calificacion_Promedio4 = tienda4['Calificación'].mean()
print(f'Calificación promedio de la cuarta tienda {calificacion_Promedio4}')
```
